```python
Q1=df2.quantile(0.25)
Q3=df2.quantile(0.75)
IQR = Q3 - Q1
print(Q1)
print(Q3)
print(IQR)

df2 =df2[~((df2<(Q1-1.5*IQR)) | (df2>(Q3+1.5*IQR))).any(axis=1)]
display(df2)
```
