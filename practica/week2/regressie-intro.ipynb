## Import

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as cp
```

## Predefine

```python
def plot_init():
    
    fig, ax = plt.subplots()
    
    return ax


def plot_add(ax, x, y):
    
    ax.plot(x, y, linewidth=2.0)


def plot_show():
    
    plt.show()


def plot_graph(x, y):
    
    ax = plot_init()
    plot_add(ax, x, y)
    plot_show()
```

## Lineair verband
### Model
huizenprijs = vastgoed offset + prijs per oppervlak x oppervlak

p = 25.000 + 5.000 x oppervlak

### Stap 1: Maak huizenprijs model

```python
set_off_value = 25000 
prijs_per_sq = 5000
total_area = np.arange(start= 50, stop = 200)
print(total_area)


total_prijs= set_off_value + (prijs_per_sq*total_area)
print(total_prijs)

plt.plot(total_area,total_prijs,c='r')
plot_show()

# noise into the data

total_prijs_noise = np.random.normal(total_prijs,55000)

plt.plot(total_area, total_prijs_noise)

plot_show()

```


### Stap 2: Bekijk interval [50, 75] m2

```python


# slicing the total data


total_area_50_75= total_area[:20]
print(total_area_50_75)
total_prijs_50_75=total_prijs_noise[:20]
print(total_prijs_50_75)
plt.plot(total_area_50_75, total_prijs_50_75, c ='r')


```

### Stap 3: Fit data

#### Use the curve fitting

https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html

```python
def function(x,a,b):
    return (b * x) + a
   


total_prijs_fit, pcos = cp.optimize.curve_fit(function,total_area_50_75,total_prijs_50_75)
print(total_prijs_fit)
new_prijs_prediction = total_prijs_fit[0]+ total_prijs_fit[1]* total_area_50_75
ax =plot_init()

plot_add(ax, total_area_50_75, total_prijs_50_75)
plt.scatter(total_area_50_75, total_prijs_50_75, c='r')
plot_add(ax, total_area_50_75, new_prijs_prediction)



plt.show()
```

## Polynomiale verbanden

### Model



massa bolmannetje = dichtheid x bolvormigheid + benen + armen + hoofd

mass_bolman = 1.03 x radius^3 + 12.4 + 8.7 + 3.2

### Stap 1: Maak model bolmannetje

```python
benen = 12.4
armen = 8.7
hoofd = 3.2
setOfPrice= 1.03

mass_bodyparts = benen+hoofd+armen


radius = np.arange( 10, 20, 0.1)
print(radius)

mass_bolman =( setOfPrice *(radius**3) )+ mass_bodyparts

print(mass_bolman)

plt.plot(radius, mass_bolman)
plt.show()


#noise into the data

mass_bolman_noise = np.random.normal(mass_bolman, 50)
print(mass_bolman_noise)
plt.plot(radius, mass_bolman_noise)
plt.show()


#slicing
radius_new = radius[:60]
mass_bolman_new = mass_bolman_noise[:60]
plt.plot(radius_new, mass_bolman_new)
plt.show()

```

### Stap 2: Fit data

```python
# Define our model

def function(x,a,b):
    return a + (b*(x**3))

# Use scipy optimize package

fit, pcos = cp.optimize.curve_fit(function, radius_new, mass_bolman_new)
print(fit)
# Create fit

mass_bolman_prediction = fit[0] + (fit[1]*(radius_new**3))

# Plot fit
ax = plot_init()
plot_add(ax, radius_new, mass_bolman_new)
plot_add(ax, radius_new, mass_bolman_prediction)
plt.show()
```
