Wat is de gemiddelde prijs voor ruwe olie over de jaren?

```python
# Zoek in de Pandas documentatie een geschikte functie
# Hint: https://pandas.pydata.org/docs/reference/series.html

# .mean()

# Voer deze functie uit op het veld 'Crude Oil' en print het resultaat

df_avg_crudeoil = df_commodity["Crude Oil"].mean()
display(df_avg_crudeoil)
```

Wanneer was de olieprijs op zijn hoogst? En op zijn laagst?

```python
# Print highest and lowest price
crudeoil_low = df_commodity[["Crude Oil"]].min()
display(crudeoil_low)

crudeoil_high = df_commodity[["Crude Oil"]].max()
display(crudeoil_high)

# Print het jaar van de hoogste en laagste prijs

oilMaxIndex = df_commodity["Crude Oil"].idxmax()
display(df_commodity["Year"][oilMaxIndex])

oilMinIndex=df_commodity["Crude Oil"].idxmin()
display(df_commodity["Year"][oilMinIndex])
```

Net als SQL kan Pandas ook queries uitvoeren en filteren.

Zoek alle jaren waarin de olieprijs tussen 50 en 80 dollar was.

```python
# Hint: filter jaren waar de prijs tussen 50 en 80 was
df_commodity[(df_commodity["Crude Oil"] > 50) & (df_commodity["Crude Oil"] < 80)]
oilprijs_50_80 = df_commodity[(df_commodity["Crude Oil"] > 50) & (df_commodity["Crude Oil"] < 80)]
display(oilprijs_50_80)
display(oilprijs_50_80["Year"])
                                                   
    
oil = df_commodity.query('`Crude Oil` > 50 and `Crude Oil` < 80')
display(oil)
```
