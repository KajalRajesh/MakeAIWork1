<h3>Exercise 1</h3>
<h4>Gegeven</h4>

```python
benelux = {"Belgium", "Netherlands", "Luxembourg"}
nordics = {"Denmark", "Finland", "Iceland", "Norway"}
baltics = {"Estonia", "Latvia", "Lithuania"}
```

<h4>Gevraagd</h4>
<ol>
    <li>Voeg de kolom <u>groups</u> toe aan de dataset <u>europeDF</u> met daarin de groep waartoe een land behoort volgens bovenstaande sets</li>
    <li>Ga na of er nog meer landengroepen zijn binnen Europa en voeg die ook toe</li>
    <li>Sla deze verrijkte dataset op in een nieuw CSV bestand</li>
</ol>

<h4>Oplossing</h4>

```python
# Oplossing
def group_add (row):
    if row["name"] in {"Belgium", "Netherlands", "Luxembourg"}:
        return "benelux"
    if row["name"] in {"Denmark", "Finland", "Iceland", "Norway"}:
        return "nordics"
    if row["name"] in {"Estonia", "Latvia", "Lithuania"}:
        return "baltics"
    return "others"
europeDF['groups'] = europeDF.apply (lambda row: group_add(row), axis=1)
europeDF.to_csv("C:/Users/Lenovo/MakeAIWork1/practica/week3/sql/csv/groups.csv")

display(europeDF)
```

```python
# het aantal landen dat onderdeel is van de Nordics

dbCursor = dbConnection.cursor()

#To find the country names(distinct gives unique values and avoids duplicate values) within group name as nordics:

#query = """ 
   # SELECT distinct(name)
    #FROM countries 
   # WHERE groups = 'nordics'
   # """    
 

query = """ 
    SELECT count(1)
    FROM countries 
    WHERE groups = 'nordics'
    """       
aantalLandenResult = dbCursor.execute(query).fetchall()
print(aantalLandenResult)
```

```python
#  het aantal Europese landen dat lid is van de Europese Unie  
query = """ 
    SELECT count(*)
    FROM countries 
    WHERE continent = 'eu'
    """   
aantalEuLandenResult = dbCursor.execute(query).fetchall()
print(aantalEuLandenResult)
```

```python
#het aantal EU-landen dat deel uitmaakt van de Nordics

query = """ 
    SELECT count(*)
    FROM countries 
    WHERE continent = 'eu' AND groups = 'nordics'
    """   
aantalEuLandenNordics = dbCursor.execute(query).fetchall()
print(type(aantalEuLandenNordics))
```
